# Frost depth in ground

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
````

````python
def HalfSpace(x, t, sig_1d, P):
    sigma = sig_1d * P**0.5
    T = np.exp(-x / sigma) * np.cos(2*np.pi / P * t - x / sigma)
    return T
````

## Temperature fluctuations

````python
Nx = 200
Nt = 36
T  = np.zeros((Nt, Nx))
x  = np.linspace(0, 5, Nx)

lam    = 1
rho_cp = 1500 * 1000
a      = lam / rho_cp
sig_1d = (24*3600 / np.pi * a)**0.5

T_mean = 10
dT_a   = 10
dT_w   = 10
P_a    = 365
P_w    =  14
for nt in range(0, Nt):
    t = 24*3600 * (Nt - (nt + 1)) / 365 
    T[nt, :] = T_mean + dT_a * HalfSpace(x, t, sig_1d=sig_1d, P=P_a) + dT_w * HalfSpace(x, t, sig_1d=sig_1d, P=P_w)
    
fig = plt.figure(figsize=(11, 5))
env_up   =   T_mean + dT_a * np.exp(-x / sig_1d / P_a**0.5) + dT_w * np.exp(-x / sig_1d / P_w**0.5) 
env_down =   T_mean - dT_a * np.exp(-x / sig_1d / P_a**0.5) - dT_w * np.exp(-x / sig_1d / P_w**0.5)
for nt in range(0, Nt):
    plt.plot(x/10, T[nt, :], color='grey', markersize=1)
plt.plot(x/10, env_up, color='green', markersize=1)
plt.plot(x/10, env_down, color='green', markersize=1)
plt.plot([0, 0.5],[0, 0], 'r-', label='')
plt.title('Temperature fluctuations in the ground due to periodic temperature fluctuations at the surface')
plt.xlabel('x / m')
plt.ylabel('T / °C')
plt.grid()
plt.show()
````

````python
print('Daily penetration depth =', f'{sig_1d:.{2}}', 'm,', ' yearly =', f'{sig_1d*365**0.5:.{2}}', 'm')
````

## Root finding to calculate the frost depth

For root finding look here for instance [Berkeley](https://pythonnumericalmethods.berkeley.edu/notebooks/chapter19.05-Root-Finding-in-Python.html) and what is a 'lambda'-function is described here [lambda](https://www.w3schools.com/python/python_lambda.asp).

````python
f = lambda x: T_mean - dT_a * np.exp(-x / sig_1d / P_a**0.5) - dT_w * np.exp(-x / sig_1d / P_w**0.5)
frost_depth = fsolve(f, 0)
print('frost depth =', f'{frost_depth[0]:.{2}}', 'm')
````
