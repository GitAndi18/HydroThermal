## 1D-Heat Transport Equation

Heat transport in homogeneous materials is described by the heat transport equation. This is also known as Fourier's 2nd law, named after the French researcher [Fourier](https://de.wikipedia.org/wiki/Joseph_Fourier) who worked intensively on the phenomenon of "heat". The equation in one dimension is

$$
  \rho \, c_p \, \dot{\vartheta}(x, t) = \lambda \, \vartheta^{\prime\prime}(x, t)
$$

In semi-infinite space, a cosine-shaped excitation at the surface results in the following solution (the derivation is done in the lecture) for the oscillating case. 

$$
  \vartheta(x, t) = \Delta \vartheta_0 \, \text{e}^{-\frac{x}{\sigma}} \, \text{cos}(\omega t - \frac{x}{\sigma})
$$

with the surface flucuations $\vartheta_0(t) = \Delta \vartheta_0 \, \text{cos}(\omega t)$ and the penetration depth $\sigma = \sqrt{\frac{T}{\pi}\,a}$. $a$ is the thermal diffusity defined as $a = \frac{\lambda}{\rho c_p}$ and $T$ is the period of the harmonic fluctuations. 

The following graph shows characteristic temperature curves in the ground for an annual period.

````python
T = 3600*24*365
w = 2*np.pi / T
Nt = 11
Nx = 50
x_max = 10
t = np.linspace(0, T, Nt)
x = np.linspace(0, x_max, Nx)
theta = np.zeros((Nt, Nx))
dtheta_0   = 10
theta_mean = 10
lam    = 1
rhocp  = 1500*1000
a      = lam / rhocp
sigma  = (T/np.pi * a)**0.5

for n, tt in enumerate(t):
    theta[n, :] = theta_mean + dtheta_0 * np.exp(-x / sigma) * np.cos(w*tt + x / sigma)

#%matplotlib ipympl
fig = plt.figure(figsize=(13, 5))
for n in range(Nt):
    plt.plot(x, theta[n, :], label=str(n))
plt.legend(loc='center right')
plt.title('Yearly Temperature Fluctuations in the Ground')
plt.xlabel('$x$ / m')    
plt.ylabel(r'$\vartheta$ /C')
plt.grid()
plt.show()
````
